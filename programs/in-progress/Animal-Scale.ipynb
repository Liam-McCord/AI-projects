```python
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (7,7))

ax = plt.axes()
ax.set_xlabel('X')

ax.set_ylabel('Y')

ax.grid()
ax.legend()
class YourScale():
    def __init__(self, name, bc_gr, tar_blu):
        self.bc_gr = bc_gr
        self.tar_blu = tar_blu
        self.name = name
    def plot_scale(self):
        ax.scatter(self.bc_gr, self.tar_blu, label=self.name)


Hannah = YourScale("Hannah", 0.9, 0.2)
Hannah.plot_scale()
```

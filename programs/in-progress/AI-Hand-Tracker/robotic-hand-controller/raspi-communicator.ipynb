```python
import socket
import time
from threading import Timer
import pandas as pd
from shared_memory_dict import SharedMemoryDict

finger_data_memory = SharedMemoryDict(name='finger-data', size=1024)

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
s.bind(('',5000))
s.listen(5)
print('Server is now running.')
clientsocket, address = s.accept()

finger_data_memory["Take-Scale"] = False

def background_controller():
    finger_seps = finger_data_memory["data"]
    print(finger_seps)
    
    message = str(finger_seps)
    clientsocket.send(bytes(message, "utf-8"))
    time.sleep(0.1)
    Timer(1, background_controller)
    
while True:
    print(f"Connection from (address) has been established")
    finger_data_memory["Take-Scale"] = True
    background_controller()


```
